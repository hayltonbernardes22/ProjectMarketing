### 3. Baseline Model 


Vamos criar um modelo inicial, é de suma importância tirar suas métricas de forma que possamos pautar os próximos resultados, com esse modelo baseline e assim comparar a qualidade de nossos resultados. 



<hr>
<br>


```
# drop ID 
marketing.drop('Unnamed: 0', axis=1, inplace=True)
marketing_2 = marketing.copy()


def baseline(marketing): 

  # label target 
  label = LabelEncoder()
  marketing['y'] = label.fit_transform(marketing[['y']])


  # tipo de variáveis 
  cat_cols = marketing.select_dtypes(include=['object']).columns 
  num_cols = marketing.select_dtypes(include=['float64', 'int64']).columns



  # encoding categóricas 
  dummies = pd.get_dummies(marketing[cat_cols], prefix=cat_cols, drop_first=True)
  marketing = pd.concat([marketing, dummies], axis=1)
  marketing.drop(cat_cols, axis=1, inplace=True)
  marketing.shape


  # split 
  X = marketing.drop('y', axis=1)
  y = marketing['y']

  X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)

  # padronização
  std = StandardScaler()
  X_train = std.fit_transform(X_train)
  X_test = std.transform(X_test)

  baseline = LogisticRegression(random_state=42)
  baseline.fit(X_train, y_train)
  y_pred = baseline.predict(X_test)
  y_proba = baseline.predict_proba(X_test)[:,1]

  print('AUC: {}'.format(roc_auc_score(y_test, y_proba).round(2)*100))
  print('AUPRC: {}'.format(average_precision_score(y_test, y_proba).round(3)*100))
  print('\n')
  metrics = print(classification_report(y_test, y_pred))

  return metrics


baseline(marketing)
```

Os resultados da Baseline de um modelo puro, foram baixissímos e devemos corrir isso através de teste com outras combinações de técnicas e algoritmos.

Objetivo é maximizar principalmente <b> Precision-Recall </b> que é nossa métrica principal neste problema.

### 4. Criação modelo

```
# tipo de variáveis 
cat_cols = marketing.select_dtypes(include=['object']).columns 
num_cols = marketing.select_dtypes(include=['float64', 'int64']).columns


# encoding categóricas 
dummies = pd.get_dummies(marketing[cat_cols], prefix=cat_cols, drop_first=True)
marketing = pd.concat([marketing, dummies], axis=1)
marketing.drop(cat_cols, axis=1, inplace=True)

X = marketing.drop('y', axis=1)
y = marketing['y']


# UnderSampling  
under = NearMiss(sampling_strategy={0:5500, 1:4787},random_state=42)
X, y = under.fit_sample(X, y)



# Validação 
KFold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


fold = 0
for train_index, test_index in KFold.split(X,y):
      fold += 1 
      print('Fold: ', fold)
      print('Treino: ', train_index.shape[0])
      print('Teste: ', test_index[0])

      # Classe balanceadas
      print('Não: {}  |  Sim: {}'.format(np.bincount(y)[0], np.bincount(y)[1]))

      # separando dados 
      X_train, X_test = X[train_index], X[test_index]
      y_train, y_test = y[train_index], y[test_index] 

      
      # pré-processamento 
      scaler = StandardScaler()
      X_train = scaler.fit_transform(X_train)
      X_test = scaler.transform(X_test)




      # Criando modelo 
      mdl = RandomForestClassifier(n_estimators=1000,  max_depth=18,random_state=42)
      mdl.fit(X_train, y_train)
      y_pred = mdl.predict(X_test)
      y_proba = mdl.predict_proba(X_test)[:,1]

      def objetcive(params):
          all_params = {**params}
          return -average_precision_score(y_test,y_pred)


      spaces = {'n_estimators': hp.randint('n_estimators', 2000),
                'max_depth': hp.randint('max_depth', 30),
                'min_samples_leaf': hp.uniform('min_samples_leaf', 1,40),
                'max_features': hp.randint('max_features', 42),
                'min_impurity_decrease': hp.uniform('min_impurity_decrease', 0.01, 0.20)}


      trials = Trials()
      best = fmin(fn=objetcive, algo=tpe.suggest, space=spaces, trials=trials, max_evals=40)


      # Métricas
      print('\n')
      print(classification_report(y_test, y_pred))
      print('-------------'*5)
      print('\n')
      auc = roc_auc_score(y_test, y_proba)
      precision = precision_score(y_test, y_pred)
      recall = recall_score(y_test, y_pred)
      auprc = average_precision_score(y_test, y_proba)
```

```
print('---- Métrica Random Forest + Nearmiss ----')
print('\n')


print('AUC: {}'.format(np.mean(auc).round(2)))
print('Precision: {}'.format(np.mean(precision).round(2)))
print('Recall: {}'.format(np.mean(recall).round(2)))      
print('Precision-Recall: {}'.format(np.mean(auprc).round(2))) 
```
